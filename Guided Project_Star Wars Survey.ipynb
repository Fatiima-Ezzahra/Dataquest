# Guided Project: Star Wars Survey

In this notebook, I'll be working on survey data provided by the team at [FiveThirtyEight](https://fivethirtyeight.com/) from 835 Star Wars fans.

## Reading the data

```python
import pandas as pd
```

```python
star_wars = pd.read_csv("star_wars.csv", 
                        encoding='ISO-8859-1')
```

We need to specify an encoding because the data set has some characters that aren't in Python's default **utf-8** encoding. You can read more about character encodings on [developer Joel Spolsky's blog](http://www.joelonsoftware.com/articles/Unicode.html).

## Exploring the data

```python
star_wars.head(10)
```

```python
star_wars.columns
```

## Removing Null RespondentID

```python
star_wars = star_wars[pd.notnull(star_wars['RespondentID'])]
```

```python
star_wars.head()
```

## Convert String values to Boolean

We'll convert the 2nd and 3rd columns values to **True**, **False** or **NaN** because they're easier to work with.

```python
yes_no = {"Yes":True,"No":False}
cols = ['Have you seen any of the 6 films in the Star Wars franchise?',
            'Do you consider yourself to be a fan of the Star Wars film franchise?']

for col in cols:
    star_wars[col] = star_wars[col].map(yes_no)
    
```

```python
star_wars.head()
```

```python
star_wars['Unnamed: 4'].value_counts(dropna=False)
```

For the next 6 columns, the respondents check off every movie they saw. Therefore, we'll change values in these columns to **True** if the value is the name of the movie and **False** if it's **NaN**. Then, we'll change the name of the column to make it easier.

```python
import numpy as np

movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}
```

```python
for c in star_wars.columns[3:9]:
    star_wars[c] = star_wars[c].map(movie_mapping)
```

```python
star_wars.head()
```

```python
star_wars = star_wars.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"
        })
```

```python
star_wars.head()
```

## Convert columns to numeric type

Next, we want to convert the next 6 columns to numeric type and rename them.

```python
n_cols = star_wars.columns[9:15]
star_wars[n_cols] = star_wars[n_cols].astype(float)
```

```python
star_wars = star_wars.rename(columns={
        "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
        "Unnamed: 10": "ranking_2",
        "Unnamed: 11": "ranking_3",
        "Unnamed: 12": "ranking_4",
        "Unnamed: 13": "ranking_5",
        "Unnamed: 14": "ranking_6"
        })
```

```python
star_wars.head()
```

## Finding the highest ranking movie

```python
rankings = star_wars.columns[9:15]

avg_rank = star_wars[rankings].mean()
```

```python
avg_rank
```

```python
%matplotlib inline

avg_rank.plot.bar()
```

So far, we cleaned up some column names to make the dataframe easier to read and work with. We cleaned up the data, converted some string data to boolean, and others to numeric to make data analysis easier.

It looks like newer movies rank better than older ones.

## Analysing how many respondents have seen each movie

```python
seen_cols = star_wars.columns[3:9]

n_seen = star_wars[seen_cols].sum()
```

```python
n_seen
```

```python
n_seen.plot.bar()
```

It looks like more respondents have seen newer movies rather than older ones. This may explain their ranking.
